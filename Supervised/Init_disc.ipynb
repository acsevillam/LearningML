> ¿Cómo saber si mi modelo reproduce bien el fenómeno?

```python
x=10*np.random.random(100)
```

```python
y=x+2*np.random.random(100)
```

```python
plt.plot(x,y,'.')
```

```python
x_1=x[:10]
y_1=y[:10]
```

```python
plt.plot(x,y,'.')
plt.plot(x_1,y_1,'.')
```

# Persona 1

```python
plt.plot(x_1,y_1,'.')
```

```python
def f(x,a,b,c,d,e,f,g,h,i,j):
    return a+b*x+c*x**2+d*x**3+e*x**4+f*x**5+g*x**6+h*x**7+i*x**8+j*x**9
def g(x,a,b,c,d,e,f,g,h,i):
    return a+b*x+c*x**2+d*x**3+e*x**4+f*x**5+g*x**6+h*x**7+i*x**8
```

```python
popt,pcov=curve_fit(f,x_1,y_1)
```

```python
pcov
```

```python
x_pred=np.linspace(0.3,9,1000)
plt.plot(x_1,y_1,'.')
plt.plot(x_pred,f(x_pred,*popt))
```

```python
popt,pcov=curve_fit(g,x_1,y_1)
```

```python
x_pred=np.linspace(0.3,9,1000)
plt.plot(x_1,y_1,'.')
plt.plot(x_pred,g(x_pred,*popt))
min_sq=0
for i in range(len(x_1)):
    plt.plot([x_1[i]]*2,[y_1[i],g(x_1[i],*popt)],color='C2')
    min_sq+=(y_1[i]-g(x_1[i],*popt))**2
print(min_sq)
```

```python
x_pred=np.linspace(0,10,1000)
plt.plot(x,y,'.')
plt.plot(x_pred,g(x_pred,*popt))
min_sq=0
for i in range(len(x)):
    plt.plot([x[i]]*2,[y[i],g(x[i],*popt)],color='C2')
    min_sq+=(y[i]-g(x[i],*popt))**2
print(min_sq)
```

```python
def r(x,a,b):
    return a*x+b
popt_r,pcov_r=curve_fit(r,x,y)
```

```python
x_pred=np.linspace(0,10,1000)
plt.plot(x,y,'.')
plt.plot(x_pred,g(x_pred,*popt))
min_sq=0
min_sq_r=0
for i in range(len(x)):
    plt.plot([x[i]]*2,[y[i],g(x[i],*popt)],color='C2')
    min_sq+=(y[i]-g(x[i],*popt))**2
    min_sq_r+=(y[i]-r(x[i],*popt_r))**2
plt.plot(x_pred,r(x_pred,*popt_r),color="C3")
print(min_sq,min_sq_r)
```
