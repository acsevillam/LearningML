```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
X,y= make_moons(100)
```

```python
print(y)
```

```python
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
gamma=8
zs=np.exp(-gamma*(X[:,0]**2+X[:,1]**2))
ax.scatter(X[:,0], X[:,1], zs, c=y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
X_plot=X[y==0]
plt.scatter(X_plot[:,0],X_plot[:,1])
X_plot=X[y==1]
plt.scatter(X_plot[:,0],X_plot[:,1])
```

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
X_pca=pca.fit_transform(X)
```

```python
X_plot=X_pca[y==0]
plt.scatter(X_plot[:,0],X_plot[:,1])
X_plot=X_pca[y==1]
plt.scatter(X_plot[:,0],X_plot[:,1])
```

```python
from sklearn.decomposition import KernelPCA
```

```python
Kpca=KernelPCA(n_components=2,kernel="rbf",gamma=15)
```

```python
X_kpca=Kpca.fit_transform(X)
```

```python
X_plot=X_kpca[y==0]
plt.scatter(X_plot[:,0],X_plot[:,1])
X_plot=X_kpca[y==1]
plt.scatter(X_plot[:,0],X_plot[:,1])
```

```python
from sklearn.datasets import make_circles
```

```python
X,y=make_circles(n_samples=1000, noise=0.1, factor = 0.2)
```

```python
X_plot=X[y==0]
plt.scatter(X_plot[:,0],X_plot[:,1])
X_plot=X[y==1]
plt.scatter(X_plot[:,0],X_plot[:,1])
```

```python
Kpca=KernelPCA(n_components=2,kernel="rbf",gamma=8)

X_kpca=Kpca.fit_transform(X)
```

```python
X_plot=X_kpca[y==0]
plt.scatter(X_plot[:,0],X_plot[:,1])
X_plot=X_kpca[y==1]
plt.scatter(X_plot[:,0],X_plot[:,1])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


zs=np.exp(-(X[:,0]**2+X[:,1]**2))

ax.scatter(X[:,0], X[:,1], zs, c=y)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
